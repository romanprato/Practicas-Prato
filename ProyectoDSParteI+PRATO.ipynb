```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
archivo = "heart.csv"  
datos = pd.read_csv(archivo)

print(datos.head())
```

```python
# Identificar valores perdidos. Este código es para analizar si hay celdas vacías o datos ausentes en las columnas del dataset.
print("Valores perdidos por columna:")
print(datos.isnull().sum())
```

```python
print(datos.describe())
```

## Distribución de edades según presencia de enfermedad cardíaca


```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='target', y='age', data=datos, hue='target', palette='coolwarm', dodge=False)
plt.title('Distribución de edades según enfermedad cardíaca')
plt.xlabel('Enfermedad cardíaca (0=No, 1=Sí)')
plt.ylabel('Edad')
plt.legend(title='Enfermedad cardíaca')
plt.show()
```

### Distribución de las edades por grupo:

Grupo 0 (sin enfermedad cardíaca):
- La mediana está en torno a los 55 años.
- Los datos están más concentrados entre los 45 y 60 años, con una menor variabilidad.
- Hay un valor atípico cercano a los 30 años (fuera del bigote inferior).

Grupo 1 (con enfermedad cardíaca):

- La mediana está más cerca de los 60 años, lo que indica que este grupo es ligeramente mayor en promedio.
- Hay mayor dispersión de las edades, con un rango intercuartílico más amplio (entre 50 y 65 años aproximadamente).

Valores extremos (atípicos):

- En el grupo 0, el valor atípico es un individuo significativamente más joven que el rango principal (alrededor de los 30 años).
- No se observan valores atípicos en el grupo 1.


Rango de edades:

- Ambas categorías tienen un rango de edades similar en términos de bigotes (desde los 40 hasta los 70+ años).

### Insights relevantes

Edad como posible factor de riesgo:
Las personas con enfermedad cardíaca (grupo 1) parecen tener una mediana de edad más alta que las personas sin enfermedad (grupo 0). Esto sugiere que la edad podría ser un factor importante asociado al riesgo de desarrollar enfermedad cardíaca.

Mayor dispersión en el grupo 1:
La variabilidad en las edades del grupo 1 indica que la enfermedad puede estar presente en un rango más amplio de edades, aunque la mayoría de los casos se concentran entre los 50 y 65 años.

Posible caso atípico en el grupo 0:
El valor atípico en el grupo 0 podría representar un caso especial de una persona joven sin enfermedad cardíaca, que tal vez tenga características únicas que expliquen su salud en comparación con la población general.

## Gráfico 2: Colesterol, tipo de dolor torácico y enfermedad cardíaca

Analiza cómo varía el colesterol según el tipo de dolor torácico y la presencia de enfermedad cardíaca.

```python
plt.figure(figsize=(10, 8))
sns.violinplot(x='cp', y='chol', hue='target', data=datos, split=True, palette='muted')
plt.title('Colesterol según tipo de dolor torácico y enfermedad cardíaca')
plt.xlabel('Tipo de dolor torácico')
plt.ylabel('Colesterol [mg/dL]')
plt.legend(title='Enfermedad cardíaca (0=No, 1=Sí)')
plt.show()
```

### Interpretación de las variables:

Eje X: Representa los diferentes tipos de dolor torácico, categorizados como 0, 1, 2 y 3.

Eje Y: Muestra los niveles de colesterol.

Colores: Diferencian entre pacientes con enfermedad cardíaca (naranja) y sin enfermedad cardíaca (azul).

### Distribución general del colesterol:

Para todos los tipos de dolor, la mayor densidad de colesterol se encuentra entre 200 y 300 mg/dL.
Existen algunos valores extremos (outliers) de colesterol muy altos (> 400 mg/dL), especialmente en los casos con enfermedad cardíaca.

### Por tipo de dolor torácico:

- Tipo 0: Este tipo de dolor muestra mayor dispersión en el colesterol, tanto en pacientes con como sin enfermedad cardíaca. La superposición de colores es alta, indicando similitudes en las distribuciones.
- Tipo 1, 2 y 3: La densidad se concentra más en un rango reducido de colesterol (aproximadamente entre 200 y 300 mg/dL), y hay una clara separación entre los grupos con y sin enfermedad cardíaca.

### Diferencias entre los grupos:

Los pacientes con enfermedad cardíaca (naranja) tienden a tener niveles de colesterol ligeramente más altos en promedio que aquellos sin enfermedad cardíaca (azul), particularmente en los tipos de dolor 1 y 2.

## Insights relevantes

### Colesterol como posible indicador de riesgo:
- Aunque ambos grupos tienen distribuciones similares en términos de colesterol, los pacientes con enfermedad cardíaca tienden a presentar una mayor densidad en niveles más altos de colesterol.
- Esto refuerza el papel del colesterol como un factor de riesgo asociado a enfermedades cardíacas.

### Tipo de dolor torácico como diferenciador:
- En tipos de dolor más específicos (1, 2 y 3), las distribuciones entre los grupos con y sin enfermedad se diferencian más, sugiriendo que ciertos patrones de dolor pueden estar relacionados con el nivel de colesterol y la presencia de enfermedad.

### Valores extremos:
- Los valores atípicos de colesterol en pacientes con enfermedad cardíaca (sobre todo en el tipo de dolor 0) pueden representar casos críticos que merecen más atención médica.

## Edad, frecuencia cardíaca máxima y género


```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='age', y='thalach', hue='sex', style='target', data=datos, palette='coolwarm', alpha=0.8)
plt.title('Relación entre edad, frecuencia cardíaca máxima y género')
plt.xlabel('Edad de los pacientes')
plt.ylabel('Frecuencia cardíaca máxima')
plt.legend(title='Sexo (0=Mujer, 1=Hombre)')
plt.grid()
plt.show()
```

El gráfico es un scatter plot que representa la relación entre la edad y la frecuencia cardíaca máxima, con diferenciación por sexo y presencia de enfermedad cardíaca:

- Eje X: Edad de los pacientes.
- Eje Y: Frecuencia cardíaca máxima alcanzada.
- Colores: Diferencian el sexo (azul para hombres, naranja para mujeres).
- Marcadores: Diferencian el estado de la enfermedad cardíaca (círculo para sin enfermedad, cruz para con enfermedad).

### Distribución General:
A medida que aumenta la edad, la frecuencia cardíaca máxima tiende a disminuir, lo que es consistente con el conocimiento médico que indica que la capacidad cardíaca puede disminuir con la edad.

### Impacto del Género:
Parece que las mujeres (naranja) tienden a tener frecuencias cardíacas máximas más altas en comparación con los hombres (azul), especialmente en las edades medias (40-60 años).

### Relación con la Enfermedad Cardíaca:
Los individuos con enfermedad cardíaca (marcadores de cruz) están presentes en todo el rango de edades y frecuencias cardíacas, pero con una concentración notable en rangos de frecuencia cardíaca más bajos comparado con aquellos sin enfermedad (círculos), especialmente en edades avanzadas.

### Interacción entre Edad, Género y Estado de Salud:
En jóvenes y de mediana edad, no parece haber una distinción clara en la frecuencia cardíaca máxima entre quienes tienen y no tienen enfermedad cardíaca.
En personas mayores, los que tienen enfermedad cardíaca (cruces) tienden a tener frecuencias más bajas, lo que puede ser un indicativo de deterioro cardíaco relacionado con la enfermedad.

- Sexo (sex): 0.696, ya que el sexo está codificado como 0 para mujeres y 1 para hombres, este promedio sugiere una mayor proporción de hombres.

Desviación estándar:

- Presión arterial sistólica (trestbps): 17.52, indica una variabilidad moderada en las mediciones de presión arterial entre los pacientes.
- Frecuencia cardíaca máxima (thalach): 23.00, sugiere que hay una variabilidad considerable en la capacidad de ejercicio entre los individuos.

Variables clínicas y su interpretación:

- Frecuencia de angina inducida por ejercicio (exang): La media es de 0.337, lo que sugiere que aproximadamente un tercio de los pacientes experimenta angina durante el ejercicio.
- Depresión del ST (oldpeak): Media de 1.07, con un máximo de 6.2, reflejando la variabilidad en la severidad de los cambios isquémicos inducidos por el ejercicio en los pacientes.
- Inclinación del segmento ST (slope): La media es cercana a 1, lo que sugiere que la mayoría de los pacientes muestra un tipo de respuesta considerada menos riesgosa en las pruebas de esfuerzo.
- Número de vasos principales (ca) observados por fluoroscopia: El 75% de los pacientes tiene 1 vaso o menos afectado, un dato relevante para evaluar la severidad de la enfermedad coronaria.

#### Pruebas estadísticas

```python
t_stat, p_val = stats.ttest_ind(
    datos[datos['target'] == 1]['age'],
    datos[datos['target'] == 0]['age']
)
print("\nT-Test entre grupos con y sin enfermedad cardíaca:")
print("T-Statistic:", t_stat, "P-Value:", p_val)
```

Estos resultados arrojan que la característica analizada difiere entre pacientes con y sin enfermedad cardíaca. Por ejemplo, hablando de la edad, podría ser que los pacientes con enfermedad cardíaca tienden a ser mayores que aquellos sin la enfermedad.

#### Correlaciones

```python
print("\nCorrelaciones entre edad, colesterol y frecuencia cardíaca máxima:")
correlaciones = datos[['age', 'chol', 'thalach']].corr()
print(correlaciones)
```

```python
# Presentación visual de correlaciones con Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(correlaciones, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Mapa de calor de correlaciones')
plt.show()
```

Correlación entre 'edad' y 'colesterol':

Coeficiente: 0.219823
Existe una correlación positiva moderada entre la edad y los niveles de colesterol. Esto sugiere que a medida que las personas envejecen, tienden a tener niveles de colesterol más altos. La correlación no es muy fuerte, pero es significativa, indicando que la edad puede ser un factor en el aumento del colesterol, aunque no es el único determinante.

Correlación 'entre edad' y 'frecuencia cardíaca máxima':

Coeficiente: -0.390227
Existe una correlación negativa moderada entre la edad y la frecuencia cardíaca máxima. Esto indica que a medida que las personas envejecen, su frecuencia cardíaca máxima tiende a disminuir. Esta es una relación más fuerte en comparación con la correlación entre edad y colesterol, lo que refleja un patrón fisiológico esperado donde la capacidad cardíaca máxima disminuye con la edad.

Correlación entre 'colesterol' y 'frecuencia cardíaca máxima':

Coeficiente: -0.021772
La correlación es prácticamente nula, indicando que no hay una relación lineal aparente entre los niveles de colesterol y la frecuencia cardíaca máxima. Esto sugiere que, en esta muestra, los niveles de colesterol no influyen directamente en la capacidad de ejercicio medida por la frecuencia cardíaca máxima.
